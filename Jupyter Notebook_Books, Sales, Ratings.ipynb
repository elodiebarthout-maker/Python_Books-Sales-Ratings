Analysis of a Books dataset found at https://www.kaggle.com/datasets/thedevastator/books-sales-and-ratings 

```python
import numpy as np
import pandas as pd

books = pd.read_csv('/kaggle/input/datasets/thedevastator/books-sales-and-ratings/Books_Data_Clean.csv')

print('Done.')
```

```python
#What kind of information this dataset contains?
books.columns
```

```python
#Checking data types and if there are any missing values
books.info()
```

```python
#Checking missing value for 'Publishing Year'
books[books['Publishing Year'].isna()]
```

```python
#Removing missing value for 'Publishing Year' as 'Book Name' is also missing
books_v2 = books.dropna(subset = 'Publishing Year', how = 'all')
```

```python
#Checking & removing missing values for 'Book Name'
books_v2[books_v2['Book Name'].isna()]
books_v3 = books_v2.dropna(subset = 'Book Name', how = 'all')
```

```python
#Checking values for language_code
books_v3['language_code'].value_counts()
```

```python
#Renaming 'language_code' values for readability
books_v3['language_code'] = books_v3['language_code'].replace({
    'eng': 'english',
    'en-US': 'english',
    'en-GB': 'english',
    'en-CA': 'english',
    'fre': 'french',
    'spa': 'spanish',
    'ara': 'arabic',
    'nl': 'dutch'
})

books_v3['language_code'].value_counts()
```

```python
#Checking missing values for language_code
books_v3[books_v3['language_code'].isna()]
```

```python
#Assuming all books with NA value for 'language_code' are written in english based on their title
books_v3['language_code'] = books_v3['language_code'].fillna('english')
books_v3[books_v3['language_code'].isna()]
```

```python
#Checking the publication window
books_v3['Publishing Year'].agg(['min', 'max'])
```

```python
#Checking if books published before year 0 are inaccuracies or valid entries
books_v3[books_v3['Publishing Year'] < 0]
```

```python
#Let's focus on modern books by removing the ones published before 1800s
books_v4 = books_v3[books_v3['Publishing Year'] >= 1800]
books_v4.info()
```

```python
#Checking if there are any duplicate records in 'Book Name' column
books_v4['Book Name'].value_counts()
```

```python
#Checking books that are present in duplicates
books_v4[books_v4['Book Name'] == 'Persepolis'] 
#Multiple tomes of that book so valid entries
```

```python
#Checking books that are present in duplicates
books_v4[books_v4['Book Name'] == 'The Awakening']
#Different writers so valid entries
```

```python
#Checking if there are any unreadable 'Book Name'
mask = books_v4['Book Name'].str.match(
    r'^(ã|æ|è|ä|é|å|ç)',
    na=False
)

print(books_v4.loc[mask, 'Book Name'])
```

```python
#Removing rows for unreadable 'Book Name'
books_v4 = books_v4[~mask]
print(books_v4.loc[mask, 'Book Name'])
```

```python
#Checking that 'Book Name' doesn't contain any other unreadable titles by looking at 
#titles that start with at least 5 non-Latin characters
mask2 = books_v4['Book Name'].str.match(
    r'^[^A-Za-zÀ-ÿ]{5,}',
    na=False
)

print(books_v4.loc[mask2, 'Book Name'])
```

```python
#Checking consistency in 'genre' column
books_v4['genre'].value_counts()
```

```python
#Renaming 'genre' values for readability
books_v4['genre'] = books_v4['genre'].replace({
    'genre fiction' : 'fiction'
})

books_v4['genre'].value_counts()
```

```python
#Checking that there are no missing values anymore
books_v4.info()
```

```python
#Droping 'gross sales' & 'sales rank' columns
books_v5 = books_v4.drop(columns = ['gross sales', 'sales rank'])
books_v5.info()
```

```python
#Creating a new 'Gross sales' column for accuracy
books_v5['Gross sales'] = books_v5['units sold'] * books_v5['sale price']
books_v5.head(10)
```

```python
#Analysis
#What is the publication period?
books_v5['Publishing Year'].agg(['min', 'max'])
```

```python
#Analysis
#Which is the most represented language?
books_v5['language_code'].value_counts()
```
